```python
def data_to_sparkdf(data):
    #Encoding "ISO-8859"
    data_v = data.decode(encoding ='ISO-8859-15')
    data_v0 = data_v
    # Clean the data
    # Step 0 
    for i in range(20):
        if(data_v0[i]=='{'):
            data_v0 = data_v0[i:]
    for i in range(20):
        if(data_v0[-i]=='}'):
            data_v0 = data_v0[:-i+1]
    # Step 1     
    data_v1 = data_v0
    data_v1 = data_v1.replace("\n", "")
    
    # Step 2
    data_v2 = data_v1.replace("},","}},")
    
    # Step 3
    patron =['\s\s','\s"','"\s','\s{']
    replace = [' ','"','"','{']
    
    data_v3 = data_v2
    for i in range(len(patron)):
        data_v3 = reg.sub(patron[i],replace[i],data_v3)

    # Step 4
    data_v4 = data_v3.replace("\",\"","\";\"")
    
    # Step 5
    data_cleaned = data_v4.split("},")

    # String to List of dictionaries
    diccionarios = []
    for fila in data_cleaned:
        #print(fila)
        keys = []
        values = []
        for pareja in fila[1:-1].split(';'):
            #print("Pareja= ",pareja)
            elems =pareja.split(':')
            #print("Elementos= ",elems)
            keys.append(elems[0][1:-1])
            values.append(elems[1][1:-1])
        diccionarios.append(dict(zip(keys,values)))

    # Schema for the new DF
    data_schema = [StructField('latitud',StringType(), True), #Tercer argumento = nullable
                   StructField('provincia', StringType(), True),
                   StructField('altitud', StringType(), True),
                   StructField('indicativo', StringType(), True),
                   StructField('nombre', StringType(), True),
                   StructField('indsinop', StringType(), True),
                   StructField('longitud', StringType(), True)                    
                  ]
    # Create and return the new DF
    return spark.createDataFrame(diccionarios,schema = StructType(data_schema))  
```

```python
df = data_to_sparkdf(data)
```

```python
def degrees_to_decimal(elem):
    elem = reg.sub('N|E','1',elem)
    elem = reg.sub('S|W','-1',elem)
    return (float(elem[6:])* (float(elem[0:2]) + float(elem[2:4])/60 + float(elem[4:6])/3600))
```

## [!] -  Descartar las estaciones que estén fuera del area de madrid ciudad

```python
import math
def haversine(p0,p1):
    p0_lat,p0_long =p0.split(',')
    
    lat1, lon1 = round(float(p0_lat),6),round(float(p0_long),6)
    lat2, lon2 = round(float(p1[0]),6),round(float(p1[1]),6)
    
    rad=math.pi/180
    dlat=lat2-lat1
    dlon=lon2-lon1
    R=6372.795477598
    a=(math.sin(rad*dlat/2))**2 + math.cos(rad*lat1)*math.cos(rad*lat2)*(math.sin(rad*dlon/2))**2
    distancia=2*R*math.asin(math.sqrt(a))
    return distancia
```

## [FUNC] - CLUSTERING

```python
def clustering(selected_k,df_estaciones):
    for i in range(0,selected_k-2): 
        #Vamos de más grupos a menos(mínimo 2). Si no es valido con 10, probamos con 9.. 
        num_cluster = selected_k - i
        #print("NUMERO DE CLUSTERS: ", num_cluster)

        #Crea la agrupación
        kmeans = KMeans(featuresCol='features',k=num_cluster,seed=1,maxIter=40)
        df_estaciones_model = kmeans.fit(df_estaciones)
        df = df_estaciones_model.transform(df_estaciones)

        #Se agrupa por predicción y se valida
        agrupado = df.select("prediction","MIDE_AIRE","MIDE_CLIMA_FINAL").groupBy("prediction").sum().orderBy("prediction")

        """ GRUPO ES VALIDO SI:
                1. El valor de valido > 0 (Hay estaciones de aire y de clima)
                2. El valor de valido == 0 & valido_con_descartables > 0 (solo hay estaciones de clima)
                    * SOLO NOS INTERESAN LAS ESTACIONES DE MEDICIÓN DE AIRE.
                    * Ese grupo luego se descarta
        """  

        agrupado = agrupado.select("prediction","sum(MIDE_AIRE)","sum(MIDE_CLIMA_FINAL)")
        #Sacamos columna "VALIDO":
        # VALIDO >0 implica que en ese grupo hay tanto estaciones de aire como de clima
        # VALIDO == 0 implica que en ese grupo NO hay estaciones de ambos tipos.
        agrupado = agrupado.withColumn("VALIDO",agrupado["sum(MIDE_AIRE)"]*agrupado["sum(MIDE_CLIMA_FINAL)"])
        #agrupado.show()

        #Sacamos la coluna "VALIDO_CON_DESCARTABLES"
        #VALIDO_CON_DESCARTABLES > 0 implica que ese grupo no es VALIDO pero solo hay estaciones de clima
        #VALIDO_CON_DESCARTABLES == 0 implica que ese grupo no es VALIDO y SI hay estaciones de aire. 
        agrupado = agrupado.withColumn("VALIDO_CON_DESCARTABLES", agrupado["sum(MIDE_CLIMA_FINAL)"]+agrupado["VALIDO"])
        #agrupado.show()

        no_validos = [int(row[0]) for row in agrupado.select("prediction").filter(agrupado["VALIDO_CON_DESCARTABLES"]== 0).collect()]
        #print("NO VALIDOS", no_validos)

        valido = (len(no_validos)==0)
        #print("ES VALIDO: ", valido)

        if(valido == 1):
            descartables = [int(row[0]) for row in agrupado.select("prediction").filter(agrupado["VALIDO"]== 0).collect()]
            agrupado.filter(~agrupado["prediction"].isin(descartables)).show()
            df = df.filter(~df["prediction"].isin(descartables))
            break
        #print("="*200)
    return df
```
